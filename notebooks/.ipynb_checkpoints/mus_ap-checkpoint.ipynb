```python
from collections import defaultdict

from collections import Counter

# import ete3
# from ete3 import PhyloTree
from Bio import SeqIO

import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 64)

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator

%matplotlib inline
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
PATH_TO_DATA = '../data/raw/mus_musculus/'
path_to_refseq = PATH_TO_DATA + 'mm10_ChrM_oneline.fasta'
path_to_genbank = PATH_TO_DATA + 'NC_005089.1.gb'
path_to_abasic_data = PATH_TO_DATA + '41467_2022_33594_MOESM26_ESM.csv'
path_to_G4Huter = PATH_TO_DATA + 'chrM_DNA_hl=1_k=25_G4Hseeked_2024-02-27.txt'
```

```python
refseq = str(next(SeqIO.parse(path_to_refseq, 'fasta')).seq)
genbank = next(SeqIO.parse(path_to_genbank, 'genbank'))

assert genbank.seq == refseq
```

```python
# BED format: 0-based positions, Start included, End excluded
abasic = pd.read_csv(path_to_abasic_data, sep=';')
abasic.head()
```

```python
DLOOP_START = 15422 # 0-based
WINDOW_SIZE = 10
```

```python
# 0-based
# collect triplets on mm10

df = pd.DataFrame([[i, x] for i, x in enumerate(refseq, 0)], columns=['site', 'nuc'])
df['down'] = [''] + list(df.nuc.values[:-1])
df['up'] = list(df.nuc.values[1:]) + ['']
df['triplet'] = df['down'].str.lower() + df['nuc'] + df['up'].str.lower()
df = df[(df.triplet.str.len() == 3) & (df.site < DLOOP_START)]
df[['site', 'down', 'nuc', 'up', 'triplet']]
```

```python
# merge AP sites data with mm10 triplets (site-specific)
merged = df.merge(abasic, right_on='Start(mm10)', left_on='site')
```

```python
merged
```

```python
g4conf = pd.read_table(path_to_G4Huter)
```

```python
# g4conf
```

```python
len(refseq) - g4conf['end'][28] + g4conf['start'][0]
```

```python
bp_untill_next_G4 = g4conf['start'].drop(0).values - g4conf['end'].drop(28).values
bp_untill_next_G4 = bp_untill_next_G4.tolist()
bp_untill_next_G4.append(375)  #375 bp between last G4 and first G4
```

```python
bp_untill_previous_G4 = g4conf['start'].drop(0).values - g4conf['end'].drop(28).values
bp_untill_previous_G4 = bp_untill_previous_G4.tolist()
bp_untill_previous_G4.insert(0, 375)
```

```python
g4conf['bp_untill_next_G4'] = bp_untill_next_G4
g4conf['bp_untill_previous_G4'] = bp_untill_previous_G4
```

```python
g4conf['G4_id'] = [*range(1,g4conf.shape[0]+1)]
```

```python
g4conf['up_win'] = pd.concat([g4conf[g4conf['bp_untill_previous_G4'] <= 100][['bp_untill_previous_G4', 'start']].diff(axis=1)['start'],\
                                 g4conf[g4conf['bp_untill_previous_G4'] > 100]['start']-100])
```

```python
g4conf['down_win'] = pd.concat([g4conf[g4conf['bp_untill_next_G4'] <= 100][['end', 'bp_untill_next_G4']].sum(axis=1), \
                               g4conf[g4conf['bp_untill_next_G4'] > 100]['end']+100])
```

```python
# g4conf.to_csv('../data/processed/g4conf_mm10.csv', index=False)
```

```python
data_up = []

for up_win, start, G4_id in zip(g4conf['up_win'], g4conf['start'], g4conf['G4_id']):
    
    count_window = start - up_win - WINDOW_SIZE + 1  # count window in upstream interval (before G4)
    number_of_win = count_window + 1                 # reverse counting of windows; the closer to G4, the smaller the number
    
    for i in range(up_win, start):
        row = []
        
        if i+WINDOW_SIZE <= start:
            window = [*range(i, i + WINDOW_SIZE)]
        else:
            break
#         print(i, window)    
        window_seq = ''
        for pos in window:
            if pos != 3107:
                window_seq += str(refseq[pos])
            else:
                window_seq += ''
#         print(pos, window_seq)

#       ref_annot[ref_annot['Pos'].isin(window)].Nuc.value_counts()
        
        number_of_win -= 1
        
        row.append(G4_id)
        row.append('up')
        row.append(number_of_win)
        row.append(window[0])
        row.append(window[-1])
        row.append(window_seq)
        
        data_up.append(row)
# print(data_up)
```

```python
data_down = []

for end, down_win, G4_id in zip(g4conf['end'], g4conf['down_win'], g4conf['G4_id']):
    
    number_of_win = 0
    
    for i in range(end, down_win):
        row = []
        
        # if i >= g4conf['end'].iloc[-1] | i <= range(len(refseq)+1)[-1]:
        #     window = [*range(i, range(len(refseq)+1)[-1]+1)]
        #     remain = WINDOW_SIZE - (range(len(refseq)+1)[-1] - i)
        #     window += [*range(range(len(refseq)+1)[0], remain)]
        
        # elif i > range(len(refseq)+1)[-1] and i <= down_win - WINDOW_SIZE :
        #     i = i - range(len(refseq)+1)[-1]
        #     window = [*range(i, i + WINDOW_SIZE)]
        
        if i+WINDOW_SIZE < down_win:
            window = [*range(i, i + WINDOW_SIZE)]
               
        else:
            break
            
        window_seq = ''
        for pos in window:
            window_seq += refseq[pos]
        # print(window[0], window_seq)

#     ref_annot[ref_annot['Pos'].isin(window)].Nuc.value_counts()

        number_of_win += 1

        row.append(G4_id)
        row.append('down')
        row.append(number_of_win)
        row.append(window[0])
        row.append(window[-1])
        row.append(window_seq)
        
        data_down.append(row)
# print(data_down)
```

```python
data = data_up + data_down
```

```python
df_win = pd.DataFrame(data=data, columns = ['G4_id', 'side', 'number_of_win', 'window_start', 'window_end', 'sequence'])
```

```python
nuc_count_win = df_win['sequence'].apply(lambda x: Counter(x))
```

```python
nuc_count_df_win = pd.DataFrame.from_records(nuc_count_win).sort_index(axis=1)
```

```python
df_win = df_win.join(nuc_count_df_win).fillna(0)
```

```python
data_within_start = []

for start, end, G4_id, sequence in zip(g4conf['start'], g4conf['end'], g4conf['G4_id'], g4conf['sequence']):
    
    number_of_win = 0
    
    for i, j in zip(range(start, end), range(len(sequence))):
        row = []
        
        if i+WINDOW_SIZE <= end:
            window = [*range(i, i + WINDOW_SIZE)]       
        else:
            break
        
        # print(i, window, WINDOW_SIZE, end)    
        window_seq = ''
        window_seq += sequence[j: j + WINDOW_SIZE]

        
        number_of_win += 1

        row.append(G4_id)
        row.append('within_start')
        row.append(number_of_win)
        row.append(window[0])
        row.append(window[-1])
        row.append(window_seq)
        
        data_within_start.append(row)
# print(data_down)
```

```python
data_within_end = []

for start, end, G4_id, sequence in zip(g4conf['start'], g4conf['end'], g4conf['G4_id'], g4conf['sequence']):
    
    number_of_win = 0
    
    for i, j in zip(range(end, start, -1), range(len(sequence))):
        row = []

        
        if i+WINDOW_SIZE > end:
            continue  
        elif (i+WINDOW_SIZE <= end):
            window = [*range(i, i + WINDOW_SIZE)]       
        else:
            continue
        # print(window)
        
        window_seq = ''
        window_seq += sequence[::-1][j-10:j+WINDOW_SIZE-10][::-1]   #window_seq from the end. works with '-10' 

        # print(i, j)
        
        number_of_win += 1

        row.append(G4_id)
        row.append('within_end')
        row.append(number_of_win)
        row.append(window[0])
        row.append(window[-1])
        row.append(window_seq)
        
        data_within_end.append(row)
# print(data_down)
```

```python
data_within_sides = data_within_start + data_within_end
```

```python
df_win_within_sides = pd.DataFrame(data=data_within_sides, 
                             columns = ['G4_id', 'side', 'number_of_win', 'window_start', 'window_end', 'sequence'])
```

```python
nuc_count_win_within_sides = df_win_within_sides['sequence'].apply(lambda x: Counter(x))
```

```python
nuc_count_df_win_within_sides = pd.DataFrame.from_records(nuc_count_win_within_sides).sort_index(axis=1)
```

```python
df_win_within_sides = df_win_within_sides.join(nuc_count_df_win_within_sides).fillna(0)
```

```python
AP_win = []
for window_start, window_end in zip(df_win['window_start'], df_win['window_end']):
    row = [*range(window_start, window_end+1)]
    AP_win.append(merged[(merged['Strand'] == '-') & (merged['site'].isin(row))].nuc.count())
```

```python
AP_win_within_sides = []
for window_start, window_end in zip(df_win_within_sides['window_start'], df_win_within_sides['window_end']):
    row = [*range(window_start, window_end+1)]
    AP_win_within_sides.append(merged[(merged['Strand'] == '-') & (merged['site'].isin(row))].nuc.count())
```

```python
df_win['Abasic'] = AP_win
```

```python
df_win_within_sides['Abasic'] = AP_win_within_sides
```
